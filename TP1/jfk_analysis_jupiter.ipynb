```python
%matplotlib inline

import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

_df = pd.read_csv('jfkrelease-2017-dce65d0ec70a54d5744de17d280f3ad2.csv', sep = ";", parse_dates = ['Doc Date'])

df=pd.DataFrame(_df)

mydata = df['Num Pages']
numNull = df['Num Pages'].isnull().sum()
print(mydata.describe())
print('my number of document with no page is of: ' + str(numNull))

docType = df['Doc Type'].nunique()
print('number of type of doc is : ' + str(docType))
docPage = df['Doc Type'].value_counts()
print(str(docPage))
agencies = df['Agency'].nunique()
print("the number of agencies is of : " + str(agencies))
agenciesDoc = df['Agency'].value_counts()
print(str(agenciesDoc))

fig, axs = plt.subplots(1,2)
agenciesDoc.plot(kind='bar', log='True', ax=axs[0])
docPage.plot(kind='bar',log='True', ax=axs[1])

plt.show()
sns.countplot(df['Agency'])
plt.show()
d=df['Doc Date']
d= pd.to_datetime(d, format='%M/%d/%Y', errors='coerce')
d.groupby(d.dt.strftime('%Y'))

sns.countplot(d)
plt.show()











```
